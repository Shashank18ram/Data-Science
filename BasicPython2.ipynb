```python
#while loop is nothing but it continue to execut untill its true once its reach its point it will compare with end number and gets out
count=0
while count<10:
  print(count)
  count=count+1
```

```python
#control loop statements
#break break statements exits the loop permanetly

for i in range(0,20):
  if i==15:
    break
  print(i)
```

```python
#nested loops means run inside a loop
for i in range(3):
  for j in range(2):
      print(f"i:{i}and j:{j}")
```
