```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('/content/Diab.csv')
df.head()
```

```python
df.info(
)
```

```python
dfScaler=df.copy()
dfScaler.drop(['Gender','Diabetic','Pat_Id'],axis=1,inplace=True)
dfScaler.head()
```

```python
dfScaler.columns
```

```python
#StandardScaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

dfZScaling=scaler.fit_transform(dfScaler)
dfZScaling
```

```python
dfScaler.columns
```

```python
from io import DEFAULT_BUFFER_SIZE
#convert this into np array
dfZScaling_df=pd.DataFrame(dfZScaling,columns=dfScaler.columns)
dfZScaling_df.head()
```

```python
#MINMAXScaling
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
df_MinMax=scaler.fit_transform(dfScaler)
df_MinMax
```

```python
df_MinMaxx=pd.DataFrame(df_MinMax,columns=dfScaler.columns)
df_MinMaxx.head()
```

```python
#ROBUSTScaling
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
```

```python
df_RobustScaling=scaler.fit_transform(dfScaler)
df_RobustScaling
```

```python
df_Robust=pd.DataFrame(df_RobustScaling,columns=dfScaler.columns)
df_Robust.head()
```

Min-Max Scaling relies entirely on the absolute minimum and maximum values of a dataset to stretch or compress data between 0 and 1.

If an extreme outlier is present, it massively inflates the range, causing all normal data points to get tightly compressed into a tiny fraction close to 0.

```python
dfEncoder=df.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
dfEncoder1=ohe.fit_transform(dfEncoder[['Gender']])
dfEncoder1
```

 OneHotEncoder returns a hidden sparse matrix to save memory, which looks like a text description instead of actual data.

Setting sparse_output=False forces it to return a standard, readable NumPy array of 0s and 1s so you can easily view it and convert it into a Pandas DataFrame.

```python
dfEncoder1=pd.DataFrame(dfEncoder1,columns=ohe.get_feature_names_out())
dfEncoder1.head()
```

```python
#label Encoding
dfLabel=df.copy()
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dfLabel['Diabetic']=le.fit_transform(dfLabel[['Diabetic']])
dfLabel.head()
```

```python
data ={
    'name': ['shashank','punith','ujwal'],
    'id':[101,103,106]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
dfoe=df.copy()
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
dfoe=oe.fit_transform(dfoe[['name','id']])
dfoe
```

```python
dfoe=pd.DataFrame(dfoe,columns=oe.feature_names_in_)
dfoe.head()
```
