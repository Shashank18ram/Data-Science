```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('/content/car_dataset_200_rows.csv')
df.head()
```

```python
plt.figure(figsize=(10,2))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
```

#0.94 hp and sp are multi correlated

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
x_HP=df[['HP']]
y_SP=df['SP']
lr=LinearRegression()
lr.fit(x_HP,y_SP)
y_pred=lr.predict(x_HP)
r2=r2_score(y_SP,y_pred)
print(round(r2,2))
```

0.88% is change of sp bcs of influcence of HP

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif_hp_sp=1/(1-r2)
print(round(vif_hp_sp,2))
```

```python
x_VOL=df[['VOL']]
y_WT=df['WT']
lr=LinearRegression()

lr.fit(x_VOL,y_WT)

y_pred=lr.predict(x_VOL)

r2=r2_score(y_WT,y_pred)
print(round(r2,2))
```

```python
VIF=1/(1-r2)
print(round(VIF,2))
```

```python
x_HP=df[['HP','SP']]
y_MPG=df['MPG']
lr=LinearRegression()
lr.fit(x_HP,y_MPG)
y_pred=lr.predict(x_HP)
r2=r2_score(y_MPG,y_pred)
print(round(r2,2))
```

```python
VIF=1/(1-r2)
print(round(VIF,2))
```
