```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('/content/experience_vs_salary.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
fig=plt.subplots(figsize=(12,5))
sns.scatterplot(x='Experience (Years)',y='Salary ($)',data=df)
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,5))
sns.boxplot(df['Experience (Years)'],color='orange',ax=axes[0])
axes[0].set_xlabel('exi')
axes[0].set_ylabel('sal')
axes[0].set_title('exvssalary')
sns.boxplot(df['Salary ($)'],color='blue',ax=axes[1])
axes[1].set_xlabel('salary')
axes[1].set_ylabel('exp')
axes[1].set_title('exp vs sal')
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
```

```python
fig=plt.subplots(figsize=(12,5))
sns.regplot(x='Experience (Years)',y='Salary ($)',data=df)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,root_mean_squared_error
```

```python
lr=LinearRegression()
x=df[['Experience (Years)']]
y=df['Salary ($)']
lr.fit(x,y)
y_pred=lr.predict(x)

r2=r2_score(y,y_pred)
rmse=np.sqrt(mean_squared_error(y,y_pred))
mse=mean_squared_error(y,y_pred)
print('R2 score',r2)
print('RMSE',rmse)
print('MSE',mse)

```
