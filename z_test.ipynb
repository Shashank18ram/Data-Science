```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.weightstats import ztest as ztest
```

```python
df=pd.read_csv('/content/bulb_life_span.csv')
df.head()
```

```python
z_scores=ztest(df['lifespan_hours'],value=1000,alternative='smaller')
print(z_scores)
```

```python
from scipy.stats import alpha
alpha=0.05
if z_scores[0] <-1.96 or z_scores[0]>1.96:
  print("we reject the null point")
else:
   print(' we fail to reject the null point')
```
