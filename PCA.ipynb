```python
df = pd.read_csv(r'/content/drive/MyDrive/Datasets/breast-cancer-wisconsin-data.csv')
df.head()
```

```python
df.info()
```

```python
X = df.drop(columns=['id','diagnosis'],axis=1)
y = df['diagnosis']
```

```python
#Scale the X features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
df_scaled = pd.DataFrame(X_scaled,columns=X.columns)
df_scaled.head()
```

```python
#import PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
df_scaled_pca = pca.fit_transform(df_scaled)
df_scaled_pca = pd.DataFrame(df_scaled_pca)
df_scaled_pca.head()
```

```python
pca.explained_variance_ratio_
```

```python
df_scaled_pca.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df_scaled_pca,y)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
#Train the model
lr.fit(X_train,y_train)
```

```python
#Predict on train and test
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
#get the classification_report
from sklearn.metrics import classification_report
print(classification_report(y_train,y_train_pred))
```
