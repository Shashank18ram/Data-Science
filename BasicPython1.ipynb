```python
#Arithematic Op
num1=int(input("enter a first number"))
num2=int(input("enter a second number"))

addition=num1+num2
subtraction=num1-num2
multiplication=num1*num2
division=num1/num2
print("addition:",addition)
print("subtraction:",subtraction)
print("multiplication:",multiplication)
print("division:",division)
```

```python
#Comparison Op means we have to use == to compare values
a=10
b=10
print(a==b)
```

```python
#Logical op>in this we have AND if both condition is true it will be true or false,OR if both condition have different condition true,NOT if both condition have different it is false
a=False
b=True
result=a and not b
print(result)
```
