```python
dfDiab=pd.read_csv("/content/diabetes.csv")
dfDiab
```

```python
mean=dfDiab["BloodPressure"].mean()
print('mean',mean)
```

```python
mode=dfDiab["BloodPressure"].mode()
print('mode',mode)
```

```python
median=dfDiab["BloodPressure"].median()
print('median',median)
```

```python
dfDiab.info()
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

dfDiab['BMI_Category'] = dfDiab['BMI'].apply(bmi_category)
```

```python
dfDiab.info()
```

```python
dfDiab.describe()
```

```python
#plottinf two different plots of a section in a datset
#axes identifies the 2 differnt plot
fig ,axes=plt.subplots(1,2,figsize=(10,5))
sns.histplot(dfDiab["BloodPressure"],kde=True,color='red',ax=axes[0])
axes[0].set_xlabel("BloodPressure")
axes[0].set_ylabel("frequency")
axes[0].set_title("BloodPressure histogram")
sns.histplot(dfDiab["Glucose"],kde=True,color='black',ax=axes[1])
axes[1].set_xlabel("Glucose")
axes[1].set_ylabel("frequency")
axes[1].set_title("Glucose histogram")

```

```python
fig ,axes=plt.subplots(1,2,figsize=(10,5))
sns.boxplot(dfDiab["BloodPressure"],color='red',ax=axes[0])
axes[0].set_xlabel("BloodPressure")
axes[0].set_ylabel("frequency")
axes[0].set_title("BloodPressure histogram")
sns.boxplot(dfDiab["Glucose"],color='black',ax=axes[1])
axes[1].set_xlabel("Glucose")
axes[1].set_ylabel("frequency")
axes[1].set_title("Glucose histogram")

```

```python
#scatter show the relationship btw twu numeric data
sns.scatterplot(dfDiab,x='Age',y='BMI',color='red')
plt.xlabel("Age")
plt.ylabel("BMI")
plt.title("scatter plot of Age and BMI")
plt.show()
```

```python
#correlations
num_points=100
x_values=np.random.rand(num_points)
y_values=np.random.rand(num_points)

#weak correlation
weak_correlation=pd.DataFrame(
    {
    'Random_X':x_values,
    'Random_Y':y_values
})
weak_correlation

```

```python
sns.scatterplot(weak_correlation,x='Random_X',y='Random_Y',color='red')
plt.xlabel("Random_X")
plt.ylabel("Random_Y")
plt.title("scatter plot of Random_X and Random_Y")
plt.show()
```

```python
weak_correlation.corr()
```

```python
dfDiab=dfDiab.drop(['Pregnancies','Glucose','BloodPressure'],axis=1)
dfDiab.head()
```
